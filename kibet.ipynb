```python
import  numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import calendar

```

```python
data=pd.read_csv(r'C:\Users\sittuk\.matplotlib\Uber Drives - .csv')
data.head()
```

```python
data.isnull().sum()
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data.dtypes
data['START_DATE*']=pd.to_datetime(data['START_DATE*'],format='%m/%d/%Y %H:%M')
data['END_DATE*']=pd.to_datetime(data['END_DATE*'],format='%m/%d/%Y %H:%M')
data.dtypes
```

```python
hour=[]
dayofweek=[]
month=[]
weekday=[]
day=[]
for y in data['START_DATE*']:
     hour.append(y.hour)
     dayofweek.append(y.dayofweek)
     month.append(y.month)
     day.append(y.day)
     weekday.append(calendar.day_name[dayofweek[-1]])

data['HOUR']=hour
data['DAYOFWEEK']=dayofweek
data['MONTH']=month
data['WEEKDAY']=weekday
data['DAY']=day
```

```python
data.head()
```

```python
data['CATEGORY*'].value_counts()
```

```python
sns.countplot(x='CATEGORY*',data=data)
```

```python
#how long do people travel with uber
data['MILES*'].plot.hist()
```

```python
#what hour do  most people take uber to their destination
hour = data['START_DATE*'].dt.hour.value_counts()
hour.plot(kind='bar',figsize=(10,5),color='pink')

plt.xlabel('HOURS')
plt.ylabel('FREQUENCY')
plt.title('Number of trip vs Hours')
```

```python
#purpose of trip
data['PURPOSE*'].value_counts().plot(kind='bar',figsize=(20,10),color='blue')
```

```python
#day that has the highest number of trips
data['WEEKDAY'].value_counts().plot(kind='bar',figsize=(10,5),color='green')
```

```python
#number of trip per each day
data['DAYOFWEEK'].value_counts().plot(kind='bar',figsize=(10,5),color='brown')
```

```python
#trip in month
data['MONTH'].value_counts().plot(kind='bar',figsize=(10,5),color='purple')
```

```python
data['START*'].value_counts().plot(kind='bar',figsize=(25,10),color='red')
```

```python
#stoping point of the trip
data['STOP*'].value_counts().plot(kind='bar',figsize=(25,10),color='black')
```
